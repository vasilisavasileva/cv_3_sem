```python
import sys

import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

### Задание 1: На занятии мы разбирали, как осуществить поворот изображения с помощью матрицы поворота. Однако мы с вами заметили, что поворот происходит относительно нуля координат, что может приводить не совсем к тому результату, который мы ожидаем получить. Задача состоит в том, чтобы повернуть изображение относительно его центра (а не начала координат).

```python
img = cv.imread("messi5.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
height, width = img.shape[:2]

from math import sin, cos, pi

angle = 180 * pi / 180

tx = width//2
ty = height//2

M = np.float32([
    [cos(angle) , -sin(angle), tx - cos(angle) * tx + sin(angle) * ty],
    [sin(angle), cos(angle), ty - sin(angle) * tx - cos(angle) * ty]])

img_rot = cv.warpAffine(img, M, (width, height))
plt.imshow(img_rot)
```

## Задание 2:
### В папке находятся изображения (все кроме messi5.jpg), которые были подвернуты различным деформациям. Используя аффинные преобразования, необходимо "выправить" эти изображения (привести в исходный вид).

```python
img_flip = cv.imread("apple.jpg")
img_flip = cv.cvtColor(img_flip, cv.COLOR_BGR2RGB)
plt.imshow(img_flip)
```

```python
height, width = img_flip.shape[:2]

sx = 1
sy = -1
ty = height
M = np.float32([
    [sx, 0, 0],
    [0, sy, ty],
]
)

img_flop = cv.warpAffine(img_flip, M, (int(width * sx), int(height * sy)))

plt.imshow(img_flop)
```

```python
img_phone = cv.imread("phone.png")
img_phone = cv.cvtColor(img_phone, cv.COLOR_BGR2RGB)
plt.imshow(img_phone)
```
